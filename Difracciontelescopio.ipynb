# Funciones de Bessel

```python


from numpy import cos,sin,pi
from pylab import *

def J(m,x):
    
    def f(teta):
        return cos(m*teta - x*sin(teta))
    
    N = 1000
    a = 0.
    b = pi
    h = (b-a)/N
    
    q = f(a) + f(b) + 4*f(b-h)
    for k in range(1,N//2):
        q += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
    
    I = h/3*q/pi
    print(I)
    
    return I

x = linspace(0,20)

plot(x,J(0,x),label='J0')
plot(x,J(1,x),label='J1')
plot(x,J(2,x),label='J2')
legend()
```

```python
from numpy import cos,sin,pi,sqrt
from pylab import *

def J(m,x):	
    
    def f(teta):
        return cos(m*teta - x*sin(teta))
    
    N = 1000
    a = 0.
    b = pi
    h = (b-a)/N
    
    s = f(a) + f(b) + 4*f(b-h)
    for k in range(1,N//2):
        s += 4*f(a + (2*k-1)*h) + 2*f(a+2*k*h)
    
    I = h/3*s/pi
    
    return I

x,y = mgrid[-1:1:100j,-1:1:100j]
r = sqrt(x**2 + y**2)
longitudeonda = 0.5
k = 2*pi/longitudeonda

I = (J(1,r*k)/k/r)**2
gray()
imshow(I,vmax=0.1/10,extent=(-1,1,-1,1))
```
